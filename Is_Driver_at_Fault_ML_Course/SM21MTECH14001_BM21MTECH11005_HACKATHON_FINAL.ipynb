```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df=pd.read_csv(r"C:\Users\Lenovo\Downloads\train.csv (1)\train.csv",parse_dates=["Crash Date/Time"])
train_df.info()
```

```python
train_df['crash_year']=train_df['Crash Date/Time'].dt.year
train_df['crash_month']=train_df['Crash Date/Time'].dt.month
train_df['crash_day']=train_df['Crash Date/Time'].dt.day
train_df['crash_Day_of_week']=train_df['Crash Date/Time'].dt.weekday
train_df['crash_hour']=train_df['Crash Date/Time'].dt.hour
```

```python
train_df.info()
```

```python
del train_df['Crash Date/Time']
del train_df['Municipality']
del train_df['Off-Road Description']
del train_df['Non-Motorist Substance Abuse']
del train_df['Circumstance']
del train_df['Related Non-Motorist']
del train_df['Location']
del train_df['Person ID']
del train_df['Vehicle ID']
```

```python
train_df.describe()
```

```python
def IQR(df, feature):
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bridge = Q1 - 1.5 * IQR
    upper_bridge = Q3 + 1.5 * IQR
    df.loc[df[feature]<=lower_bridge, feature] = lower_bridge
    df.loc[df[feature]>=upper_bridge, feature] = upper_bridge
```

```python
IQR(train_df,'Vehicle Year')
```

```python
train_df.describe()
```

```python
numerical_feature = [feature for feature in train_df.columns if train_df[feature].dtype != 'object']
categorical_feature = [feature for feature in train_df.columns if train_df[feature].dtype == 'object']
discrete_feature = [feature for feature in numerical_feature if len(train_df[feature].unique()) <= 31]
continuous_feature = [feature for feature in numerical_feature if feature not in discrete_feature]
print("Numerical features are:", numerical_feature)
print("Categorical features are:", categorical_feature)
print("Discrete features are:", discrete_feature)
print("Continuous features are:", continuous_feature)
```

```python
for feature in categorical_feature:
    train_df[feature] = train_df[feature].astype('|S')
```

```python
train_df.info()
```

```python
def labEnc(df, feature):
    from sklearn.preprocessing import LabelEncoder
    labelEncoder = LabelEncoder()
    df[feature] = labelEncoder.fit_transform(list(df[feature].values))
```

```python
labEnc(train_df,'Report Number')
labEnc(train_df,'Local Case Number')
labEnc(train_df,'Agency Name')
labEnc(train_df,'ACRS Report Type')
labEnc(train_df,'Road Name')
labEnc(train_df,'Route Type')
labEnc(train_df,'Cross-Street Type')
labEnc(train_df,'Cross-Street Name')
labEnc(train_df,'Collision Type')
labEnc(train_df,'Weather')
```

```python
labEnc(train_df,'Surface Condition')
labEnc(train_df,'Light')
labEnc(train_df,'Traffic Control')
labEnc(train_df,'Driver Substance Abuse')
labEnc(train_df,'Injury Severity')
labEnc(train_df,'Drivers License State')
labEnc(train_df,'Vehicle Damage Extent')
labEnc(train_df,'Vehicle First Impact Location')
labEnc(train_df,'Vehicle Second Impact Location')
labEnc(train_df,'Vehicle Body Type')
```

```python
labEnc(train_df,'Vehicle Movement')
labEnc(train_df,'Vehicle Continuing Dir')
labEnc(train_df,'Vehicle Going Dir')
labEnc(train_df,'Driverless Vehicle')
labEnc(train_df,'Parked Vehicle')
labEnc(train_df,'Vehicle Make')
labEnc(train_df,'Vehicle Model')
labEnc(train_df,'Equipment Problems')
```

```python
label= train_df['Fault']
data=train_df.drop(columns=['Fault'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(data,label,test_size=0.4)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
RF=RandomForestClassifier(n_estimators=500,criterion='entropy')
RF.fit(X_train,Y_train)
RF.score(X_test,Y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
GB=GradientBoostingClassifier()
GB.fit(X_train,Y_train)
GB.score(X_test,Y_test)
```

```python
from xgboost import XGBClassifier
XG = XGBClassifier()
XG.fit(X_train, Y_train)
y_pred = XG.predict(X_test)
predictions = [round(value) for value in y_pred]
accuracy = accuracy_score(Y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
from lightgbm import LGBMClassifier
lgbm=LGBMClassifier()
lgbm.fit(X_train,Y_train)
lgbm.score(X_test,Y_test)
```
